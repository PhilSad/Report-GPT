```python
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
import json
import datetime
```

```python
PROMPT_TEMPLATE = """Here is my current report:'''
{report}
'''
----
Here is the new information:
'''
{new_info}
'''
----
You should update the report with the new information. Here are the instructions:
- Only output the updated report.
- You may improve the existing report.
- It should be concise and clear with lots of title, subtitles, and bullet points.
- Make sure the report is up to date with the new information.
- You should synthesize the information in the report
- You should write in markdown format.
- You should always include the source of the information right after the information. 
- The source number is indicated by the number at the start with square brackets.
- If an information is already in the report, you should add the source number to the source of this information in the report.

Take a deep breath and have fun :)"""
```

```python
chat = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.2)
prompt = PromptTemplate.from_template(PROMPT_TEMPLATE)
```

```python
def split_list(lst, n):
    # Using list comprehension to generate chunks
    return [lst[i:i + n] for i in range(0, len(lst), n)]
data = json.load(open("../data getter/yelp/yelp_amy.json"))
data = sorted(data, key=lambda x: datetime.datetime.strptime(x['date'], "%b %d, %Y"), reverse=False)
data = data[:30]

data = [f"[{i+1}] - {d['comment']}" for i, d in enumerate(data)]
chunks = split_list(data, 3)
joined = ["\n\n".join([f"* {c}" for c in chunk]) for chunk in chunks]
print(joined[0])
```
